importing libararies

```python
import pandas as pd
import numpy as np
import os
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
zip_path = tf.keras.utils.get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)
csv_path, _ = os.path.splitext(zip_path)
```

```python
df = pd.read_csv(csv_path)
```

```python
df
```

```python
df = df[5::6]
df.head()
```

make date & Time as index

```python
df.index = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
df[:26]
```

Plot the T (degC) column

```python
temp = df['T (degC)']
temp.plot()
```

Vectorization of X & y for Time Series Data

```python
# [[[1], [2], [3], [4], [5]]] [6]
# [[[2], [3], [4], [5], [6]]] [7]
# [[[3], [4], [5], [6], [7]]] [8]

def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)
  return np.array(X), np.array(y)
```

Generating and Checking Shapes of X and y with Sliding Window

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(temp, WINDOW_SIZE)
X1.shape, y1.shape
```

```python
X1
```

```python
y1
```

Splitting Data into Training, Validation, and Test Sets

```python
X_train1, y_train1 = X1[:60000], y1[:60000]
X_val1, y_val1 = X1[60000:65000], y1[60000:65000]
X_test1, y_test1 = X1[65000:], y1[65000:]
```

```python
X_train1.shape, y_train1.shape, X_val1.shape, y_val1.shape, X_test1.shape, y_test1.shape
```

Setting Up TensorFlow Keras Model and Training Configuration

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam
```

Defining and Summarizing a Sequential LSTM Model

```python
model1 = Sequential()
model1.add(InputLayer((5, 1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))
model1.summary()
```

Configuring Training Settings and Metrics for Model1

```python
cp1 = ModelCheckpoint('model1/', save_best_only=True)
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics = [RootMeanSquaredError()])
```

```python
model1.fit(X_train1, y_train1, validation_data=(X_val1,y_val1), epochs=10, callbacks=[cp1])
```

```python
X_train1
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('model1/')
```

```python
train_predictions = model1.predict(X_train1).flatten()
train_result = pd.DataFrame(data={'train predictions':train_predictions, 'Actuel': y_train1})
train_result
```
