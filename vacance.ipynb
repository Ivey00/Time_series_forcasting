```python
import pandas as pd
import numpy as np

# Créer une plage de dates pour la période souhaitée
dates = pd.date_range(start="2021-01-02", end="2022-09-30")

# Créer un DataFrame avec une colonne 'vacance' initialisée à False
data = pd.DataFrame(dates, columns=['Date'])
data['is_holiday'] = False

# Définir les périodes de vacances pour 2021-2022
vacances = [
    ("2021-10-23", "2021-11-07"),  # Toussaint
    ("2021-12-25", "2022-01-02"),  # Noël
    ("2022-04-15", "2022-05-01"),  # Printemps + good friday
    ("2022-07-06", "2022-09-02"),  # Été

    # Vacances 2020-2021 pour comparer les périodes de l'année précédente
    ("2021-01-02", "2021-01-02"),  # Noël
    ("2021-04-16", "2021-05-01"),  # Printemps
    ("2021-07-06", "2021-09-02"),  # Été
    # ADDITIONAL
    ("2021-04-02", "2021-04-02"),  # good friday
    ("2021-04-05", "2021-04-05"),  #Easter Monday
    ("2021-05-08", "2021-05-08"), #V-E day
    ("2021-05-13", "2021-05-13"),  #Ascension day
    ("2021-05-24", "2021-05-24"),  #Whit Monday
    ("2021-11-01", "2021-11-01"), #All Saints's day
    ("2021-11-11", "2021-11-11"), #Armistice day
    ("2021-05-30", "2021-05-30"), #Mother's day
    ("2021-06-20", "2021-06-20"), #Father's day
    ("2022-04-18", "2022-04-18"),  #Easter Monday
    ("2022-05-08", "2022-05-08"), #V-E day
    ("2022-05-26", "2022-05-26"),  #Ascension day
    ("2022-05-29", "2022-05-29"), #Mother's day
    ("2022-06-19", "2022-06-19"), #Father's day
    ("2022-06-06", "2022-06-06"),  #Whit Monday
]


# Mettre à jour la colonne 'vacance' pour les périodes définies
for start_date, end_date in vacances:
    data.loc[(data['Date'] >= pd.to_datetime(start_date)) & (data['Date'] <= pd.to_datetime(end_date)), 'is_holiday'] = True

# Afficher le DataFrame
data.head(30)
```
